## Discussion
### Hamiltonian path problem is a classical NP-complete problem, where the goal is to find a path through all vertices without going through any vertices more than once. 
#### I tried to find a polynomial time solution to this problem. After consulting many papers using DFSs and its variations, I designed a dynamic tree to simulate DFS and hope to reach better performance. In the second part of the algorithm, the idea is to merge left-out nodes into the path. However, I realized that this approach does not have enough information to decide a deterministic merging and replanning algorithm of the path.
#### To be exact, since we cannot expand the tree (if so it is exponential time), the only information we can rely on is the local neghboring of the vertices. Thus, when we face a uncertain choice, we have no deterministic answer for the choice. I do not currently have a good way of solving this uncertainty.
